```python
# List creation
lst = [10, 20, 30, 40]

# Display list
print("Initial list:", lst)

# Append (add at end)
lst.append(50)
print("After append:", lst)

# Insert (add at position)
lst.insert(2, 25)
print("After insert:", lst)

# Remove (remove specific element)
lst.remove(20)
print("After remove:", lst)

# Pop (remove by index, default last)
lst.pop()
print("After pop:", lst)

# Index (find position)
print("Index of 30:", lst.index(30))

# Length of list
print("Length:", len(lst))

# Sort list
lst.sort()
print("After sort:", lst)

# Reverse list
lst.reverse()
print("After reverse:", lst)

# Copy list
new_list = lst.copy()
print("Copied list:", new_list)

# Clear list
lst.clear()
print("After clear:", lst)
```

```python
# Tuple creation
t = (10, 20, 30, 40, 20)

# Display tuple
print("Tuple:", t)

# Access element
print("First element:", t[0])

# Length of tuple
print("Length:", len(t))

# Count (how many times an element appears)
print("Count of 20:", t.count(20))

# Index (position of element)
print("Index of 30:", t.index(30))

# Slicing
print("Slice (1 to 3):", t[1:4])

# Concatenation
t2 = (50, 60)
t3 = t + t2
print("After concatenation:", t3)

# Repetition
print("After repetition:", t * 2)

# Membership
print("Is 40 in tuple?", 40 in t)

# Traversing tuple
print("Tuple elements:")
for i in t:
    print(i)
```

```python
# Set creation
s = {10, 20, 30, 40}

# Display set
print("Initial set:", s)

# Add element
s.add(50)
print("After add:", s)

# Update (add multiple elements)
s.update([60, 70])
print("After update:", s)

# Remove element
s.remove(20)
print("After remove:", s)

# Discard element (no error if not present)
s.discard(100)

# Pop (removes random element)
s.pop()
print("After pop:", s)

# Length of set
print("Length:", len(s))

# Membership
print("Is 30 in set?", 30 in s)

# Set operations
a = {1, 2, 3}
b = {3, 4, 5}

print("Union:", a | b)
print("Intersection:", a & b)
print("Difference (a - b):", a - b)

# Clear set
s.clear()
print("After clear:", s)
```

```python
# Dictionary creation
d = {"name": "Alex", "age": 20, "city": "Delhi"}

# Display dictionary
print("Initial dictionary:", d)

# Access value using key
print("Name:", d["name"])

# Add new key-value pair
d["course"] = "Python"
print("After adding new item:", d)

# Update value
d["age"] = 21
print("After updating age:", d)

# Remove item using pop()
d.pop("city")
print("After pop:", d)

# Remove last inserted item
d.popitem()
print("After popitem:", d)

# Get value safely (no error if key not present)
print("Phone:", d.get("phone", "Not Found"))

# Keys, values, items
print("Keys:", d.keys())
print("Values:", d.values())
print("Items:", d.items())

# Length of dictionary
print("Length:", len(d))

# Membership (checks key)
print("Is 'name' in dictionary?", "name" in d)

# Traversing dictionary
print("Dictionary elements:")
for key, value in d.items():
    print(key, ":", value)

# Clear dictionary
d.clear()
print("After clear:", d)
```
